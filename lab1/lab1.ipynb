1. Создайте матрицу $3 × 3 $ со значениями от 0 до 8.

2. Создайте матрицу шахматной доски  $8 × 8 $, заполненную 'B' и 'W', используя функцию tile.

5. Создайте случайный вектор размером 15 и замените максимальное значение на $NaN$.


```python
vector = np.random.rand(15)
max_index = np.argmax(vector)
vector[max_index] = np.nan
print(vector)
```

7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
decart_coords = np.random.rand(15, 2)
x = decart_coords[:, 0]
y = decart_coords[:, 1]

r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

polar_coords = np.column_stack((r, theta))

print("Декарт:")
print(decart_coords)
print("Поляр:")
print(polar_coords)
```

8. По двум массивам $x$ и $y$ постройте матрицу Коши $C$
 $(C_{ij} =\frac{1}{x_i - y_j})$

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
C = 1 / np.subtract.outer(x, y)
print(C)
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

```python
matrix = np.random.randint(0, 5, (10, 3))

print("Исх матр:")
print(matrix)
unique_rows = matrix[np.apply_along_axis(lambda row: len(np.unique(row)) == len(row), 1, matrix)]
print("Стр с уник зн:")
print(unique_rows)
```

10. Извлеките уникальные строки из двумерного массива.


```python
a = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [1, 2, 3],
    [7, 8, 9],
    [4, 5, 6]
])

unique_rows = np.unique(a, axis=0)

print("Оригинальный массив:")
print(a)
print("\nУникальные строки:")
print(unique_rows)
```

# Найдите ответы на вопросы с помощью предложенного датасета:

Файл с датасетом - `olympics_dataset.csv`

```python
import pandas as pd
```

```python
df = pd.read_csv('olympics_dataset.csv')
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
summer_olympics = df[df['Season'] == 'Summer']
unique_athletes = summer_olympics['Name'].nunique()
print(unique_athletes)
```

2. Найдите участника с наибольшим количеством медалей

```python
medals = df[df['Medal'] != 'No medal']

medal_count = medals['Name'].value_counts()

top_athlete = medal_count.idxmax()
top_athlete_medals = medal_count.max()

print(f"{top_athlete} - {top_athlete_medals}")
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
sport_participants = df.groupby('Sport')['Name'].nunique()

most_popular_sport = sport_participants.idxmax()
most_popular_count = sport_participants.max()

print(f"{most_popular_sport} - {most_popular_count}")
```

4. Кто был победителем медального зачёта в 2004 году?

```python
medals_2004 = df[(df['Year'] == 2004) & (df['Medal'] != 'No medal')]
medal_counts_2004 = medals_2004.groupby('NOC')['Medal'].count()

winner_country = medal_counts_2004.idxmax()
winner_count = medal_counts_2004.max()

print(f"{winner_country} - {winner_count}")
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
usa_team = df[df['Team'] == 'United States']

usa_medals = usa_team[usa_team['Medal'] != 'No medal']

usa_medal_count_by_sport = usa_medals['Sport'].value_counts()

top_sport_usa = usa_medal_count_by_sport.idxmax()
top_sport_medals_usa = usa_medal_count_by_sport.max()

print(f"{top_sport_usa} - {top_sport_medals_usa}")
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python

import matplotlib.pyplot as plt
summer_olympics = df[df['Season'] == 'Summer']

gender_count_by_year = summer_olympics.groupby(['Year', 'Sex'])['Name'].count().unstack()

gender_count_by_year.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Количество мужчин и женщин, участвовавших в Летних Олимпийских играх каждый год')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол')
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = df[df['Medal'] == 'Gold']

gold_medal_count_by_country = gold_medals['Team'].value_counts()

top_10_countries = gold_medal_count_by_country.head(10)

top_10_countries = top_10_countries.sort_values(ascending=True)
top_10_countries.plot(kind='bar', figsize=(12, 8), color='gold')
plt.title('Топ-10 стран с наибольшим количеством золотых медалей')
plt.xlabel('Страна')
plt.ylabel('Количество золотых медалей')
plt.xticks(rotation=45)
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_team = df[df['Team'] == 'Soviet Union']

medal_count_by_year = ussr_team[ussr_team['Medal'] != 'No medal'].groupby(['Year', 'Medal'])['Medal'].count().unstack(fill_value=0)

medal_count_by_year.plot(kind='bar', stacked=True, figsize=(12, 8), color=['gold', 'silver', 'brown'])
plt.title('Кол-во медалей, заработанных СССР по годам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.legend(title='Медаль')
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
chinaTeam = df[df['Team'] == 'China']
chinaMedals = chinaTeam[chinaTeam['Medal'] != 'No medal']
numOfMenAndWomByYear = chinaMedals.groupby(['Year','Sex']).size().unstack()

numOfMenAndWomByYear.plot(kind='bar',figsize=(10,6))
plt.title('Количество медалей, завоёванных спортсменами и спортсменками команды Китая')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45, ha='right')
plt.show()
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

```python
summer_olympics = df[df['Season'] == 'Summer']

medalists = summer_olympics[summer_olympics['Medal'] != 'No medal']['Name'].unique()

no_medals = summer_olympics[~summer_olympics['Name'].isin(medalists)]

no_medals_count_by_country = no_medals['Team'].value_counts()

top_10_countries_no_medals = no_medals_count_by_country.head(10).sort_values(ascending=True)

# Строим график
top_10_countries_no_medals.plot(kind='bar', figsize=(12, 8), color='blue')
plt.title('Топ-10 стран с наибольшим количеством участников, не получивших никаких медалей на Летних Олимпийских играх')
plt.xlabel('Страна')
plt.ylabel('Количество участников')
plt.xticks(rotation=45)
plt.show()
```
